```python
import requests
from bs4 import BeautifulSoup
import csv

# URL to scrape
URL = "http://www.values.com/inspirational-quotes"

# Sending a GET request to the URL
r = requests.get(URL)

# Parsing the content with BeautifulSoup
soup = BeautifulSoup(r.content, 'html.parser')

# Initialize the list to store quotes
quotes = []

# Find the container holding the quotes
table = soup.find('div', attrs={'id': 'all_quotes'})

# Loop through each quote and extract the information
for row in table.findAll('div', attrs={'class': 'col-6 col-lg-3 text-center margin-30px-bottom sm-margin-30px-top'}):
    quote = {}
    quote['theme'] = row.h5.text
    quote['url'] = row.a['href']
    quote['img'] = row.img['src']
    quote['lines'] = row.img['alt'].split(" #")[0]
    quote['author'] = row.img['alt'].split(" #")[1]
    quotes.append(quote)

# Save the quotes to a CSV file
filename = 'inspirational_quotes.csv'
with open(filename, 'w', newline='') as f:
    w = csv.DictWriter(f, ['theme', 'url', 'img', 'lines', 'author'])
    w.writeheader()
    for quote in quotes:
        w.writerow(quote)
```

```python
import csv

# Assuming you've already written the quotes to 'inspirational_quotes.csv'
filename = 'inspirational_quotes.csv'

# Read and print the first 5 lines of the CSV file
with open(filename, 'r') as f:
    reader = csv.reader(f)

    # Print the header
    header = next(reader)
    print(header)

    # Print the first 5 rows, but handle cases where there are fewer than 5 rows
    for i, row in enumerate(reader):
        if i < 5:
            print(row)
        else:
            break
```
